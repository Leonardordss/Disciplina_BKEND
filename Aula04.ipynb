```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imigrantes_canada.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#Tratamento de dados
df.set_index(['País'], inplace=True)
anos = list(map(str,range(1980,2013)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
argentina = df.loc['Argentina', anos]
argentina
argentina_dic={
    'ano':argentina.index.tolist(),
    'imigrantes': argentina.values.tolist()
}
```

```python
#Criando o dataframe
df_brasil = pd.DataFrame(brasil_dic)
df_brasil
```

```python
df_brasil.info()
```

```python
#Criando o dataframe
df_argentina = pd.DataFrame(argentina_dic)
df_argentina
```

```python
df_argentina.info()
```

```python
plt.plot(df_brasil['ano'], df_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.xlabel("Anos",fontsize=14)
plt.ylabel("Número de Imigrantes", fontsize=14)
plt.title('Imigração do Brasil para o Canadá', fontsize=18)
plt.show()

```

```python
plt.plot(df_argentina['ano'], df_argentina['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.xlabel("Anos",fontsize=14)
plt.ylabel("Número de Imigrantes", fontsize=14)
plt.title('Imigração da Argentina para o Canadá', fontsize=18)
plt.show()
```
